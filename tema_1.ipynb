```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
```

### Funciones de Pertenencia

```python
# triangle
x = np.arange(0, 10.5, 0.5)
y = fuzz.trimf(x, [3, 4, 6])
plt.grid(True)
plt.plot(x, y)
```

```python
# trapezoidal
x = np.arange(0, 10.5, 0.5)
y = fuzz.trapmf(x, [2, 3, 7, 9])
plt.grid(True)
plt.plot(x, y)
```
