```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
```

### Funciones de Pertenencia

```python
# triangle
x = np.arange(0, 10.5, 0.1)
y = fuzz.trimf(x, [3, 4, 6])
plt.grid(True)
plt.plot(x, y)
```

```python
# trapezoidal
x = np.arange(0, 10.5, 0.5)
y = fuzz.trapmf(x, [2, 3, 7, 9])
plt.grid(True)
plt.plot(x, y)
```

```python
# S
x = np.arange(0, 10, 0.5)
y = fuzz.smf(x, 2, 8)
plt.grid(True)
plt.plot(x, y)
```

```python
#gaussiana
x = np.arange(0, 10.5, 0.1)
y = fuzz.gaussmf(x, 5, 1)
plt.grid(True)
plt.xticks(np.arange(0, 11, 1)) 
plt.yticks(np.arange(0, 1.1, 0.1))
plt.plot(x, y)
```

```python
#sigmoide
x = np.arange(0, 10.5, 0.1)
y1 = fuzz.sigmf(x, 5, -1) #el -1 indica el sentido
y2 = fuzz.sigmf(x, 5, 2)
plt.grid(True)
plt.xticks(np.arange(0, 11, 1)) 
plt.yticks(np.arange(0, 1.1, 0.1))
plt.plot(x, y1)
plt.plot(x, y2)
```
